```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
```

```python
def JacobiRelajado(A,b,x0, w, itmax=1000, tolerancia= 1e-9):
    

    x = x0.copy()
    
    itF = 0
    
    
    for it in range((itmax)):
        
        xn = np.zeros_like(x0)
        
        for i in range(len(A[0])):
            
            s1 = np.dot(A[i,:i], xn[:i])
            
            s2 = np.dot(A[i,i+1:], x[i+1:])
            
            s = s1+s2
            
            xn[i] = (1-w)*x[i] + (w*(b[i] - s)/A[i,i])
            
        residuo = np.linalg.norm(np.dot(A,x)-b)
        if np.allclose(residuo,tolerancia,atol=tolerancia):
            itF = it
            break
        x = xn
    return x,itF
```

```python
A = np.array([[3.,-1.,-1.],[-1.,3.,1.],[2.,1.,4.]])
b = np.array([1.,3.,7.])
x0 = np.array([0.,0.,0.])
```

```python
w = np.linspace(0.5,1,200)
It = []


ItM = 1000
WM = 1


for i in range(len(w)):
    x,ItF = JacobiRelajado(A,b,x0,w[i])
    It.append(ItF)
    print(ItF,x,w[i])
    
    if ItF<=ItM:
        ItM = ItF
        WM = w[i]
```
