```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
from tqdm import tqdm
```

```python
d = np.linspace(-1,1,1000)
n = 20
```

```python
E = sym.Symbol('E',real=True)
x = sym.Symbol('x',real=True)
```

```python
PolHermit = []

for j in range(0,n+1):
    a = sym.exp(-(E**2))
    H = (-1)**j*sym.exp(E**2)*sym.diff(a,E,j,evaluate=True)
    

    #H2 = sym.diff(H2,E,n,evaluate=True)
    
    #H *= H2
    
    PolHermit.append(H)
    
PolHermit #aca tenemos los polinomios de hermite empezando desde n = 0 hasta n = 20
```

```python
def Function(fun):
    return fun
```

```python
def Derivative(fun,d,h=1e-4):
    return (fun(d+h)-fun(d-h))/(2*h)
```

```python
def GetNewtonRaphson(f,df,xn,itmax = 100, precision=1e-5):
    
    error = 1
    it = 0
    
    while error > precision and it <= itmax:
        
        try:
            
            xn1 = xn - f(xn)/df(f,xn)
            
            error = np.abs(f(xn)/df(f,xn))
            
            #print(error)
            
        except ZeroDivisionError:
            
            print('Division por cero')
            
        it += 1
        xn = xn1
    
    if it == itmax:
        False
    else:
        return xn
```

```python
def GetAllRoots(x,tolerancia=8):
    
    Roots = np.array([])
    
    for i in x:
        
        root = GetNewtonRaphson(fun,Derivative,i)
        
        if root != False:
            
            croot = np.round(root, tolerancia)
            
            if croot not in Roots:
                Roots = np.append(Roots,croot)
                
    Roots.sort()
        
    return Roots
```
