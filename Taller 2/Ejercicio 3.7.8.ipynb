```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sqrt(np.tan(x))
```

```python
def ForwardDer(f, x, h):
    d = -3*f(x) + 4*f(x+h) - f(x+2*h)
    
    return d/(2*h)
```

```python
def CentralDer(f, x, h):
    return (f(x+h)-f(x-h))/(2*h)
```

```python
X, h = np.linspace(0.1, 1.1, 101, retstep = True)
```

```python
Y_for = ForwardDer(f, X, h)
```

```python
plt.plot(X, Y_for)
```

```python
Y_cen = CentralDer(f, X, h)
```

```python
plt.plot(X, Y_cen)
```

```python
def AnaliticalDer(x):
    
    return 1/((np.square(np.cos(x)))*2*np.sqrt(np.tan(x)))
```

```python
Y_ana = AnaliticalDer(X)
```

```python
plt.plot(X, Y_ana)
```

```python
plt.plot(X, Y_for, color="red")
plt.plot(X, Y_cen, color="blue")
plt.plot(X, Y_ana, color="purple")
plt.show()
```

```python
plt.plot(X,np.abs(Y_ana-Y_for),label='Right Derivative')
plt.plot(X,np.abs(Y_ana-Y_cen),label='Central Derivative')
plt.yscale('log')
plt.legend()
```

Podemos ver como en este intervalo la derivada central resulta ser más precisa que esta nueva versión para la derivada progresiva. En general, vemos como la derivada central se aproxima más a su error esperado de O(h^2) (como h=0.01=10^-2, este error sería de 10^-4, como en el gráfico). No obstante, el error de la derivada progresiva se mantuvo un poco por encima del de la derivada central, aunque, en su mayoría, en el mismo orden de magnitud. En síntesis, podemos concluir que, a pesar de que la derivada central resulte ser un poco más precisa, el error de ambas se mantiene, en general, en el mismo orden de magnitud.
