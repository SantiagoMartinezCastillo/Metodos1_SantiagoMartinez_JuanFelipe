```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
```

```python
x = np.linspace(-4,4,25)
y = np.linspace(-4,4,25)
angulo = np.linspace(0, 2 * np.pi, 360)
```

```python
def potencial(x,y,r=2,V=2):
    phi = 0
    try:
        phi = V*x*(1-(r**2/(x**2+y**2)))
    except:
        return phi
    
    return phi
```

```python
def CentralDerivativex(f,x,y,h):
    
    d = 0.
    
    if h != 0:
        d = (f(x+h,y) - f(x-h,y))/(2*h)
        
    return d
def CentralDerivativey(f,x,y,h):
    
    d = 0.
    
    if h != 0:
        d = (f(x,y+h) - f(x,y-h))/(2*h)
        
    return d
```

```python
posx = np.zeros((1,25))
posy = np.zeros((1,25))
Vx = np.zeros((25,25))
Vy = np.zeros((25,25))

for i in range(0,len(x)):
    posx[0,i] = x[i]
    for j in range(0,len(y)):
        posy[0,j] = y[j]
        F = potencial
        Vx[i][j] = CentralDerivativex(F,x[i],y[j],1e-3)
        Vy[i][j] = (CentralDerivativey(F,x[i],y[j],1e-3))*-1
          
```

```python
fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot()
radio = 2

for i in range(0,len(x)):
    for j in range(0,len(y)):
        radX = x[i]
        radY = y[j]
        normRad = np.sqrt((radX)**2+(radY)**2)
        if normRad > 2:
            ax.quiver(x[i],y[j],Vx[i,j],Vy[i,j],color ="b")
            
plt.plot(radio*np.cos(angulo),radio*np.sin(angulo),linewidth=4,color="black")
```
