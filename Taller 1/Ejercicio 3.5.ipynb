```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
```

```python
def Function(x):
    return np.sin(x)
def EDerivative(x):
    return np.cos(x)
def cuartaDev(x):
    return (np.sin(x))
```

```python
x = np.linspace(0,2*np.pi,1000)
h = x[1] - x[0]
```

```python
f = Function(x)
```

```python
def CentralDerivative(f,x,h):
    
    d = 0.
    
    if h != 0:
        d = (f(x+h) - f(x-h))/(2*h)
        
    return d
```

```python
def CuartaCD(f,x,h,i=0):
    
    while i < 4:
        
        i += 1
    
        d = 0.

        if h != 0:
            d = (f(x+h) - f(x-h))/(2*h)
            
        return CuartaCD(f,d,h,i)
        
    return d
```

```python
def DervidadPorDef(f,x,h):
    
    yarr = np.zeros((1,len(f)))
    
    
    for i in range(2,len(f)-3):
    
        yarr[0,i] = (f[i+2] - 4*f[i+1] + 6*f[i] - 4*f[i-1] + f[i-2])/h**4
        
    return(yarr)
    
    
```

```python
plt.scatter(x,DervidadPorDef(f,x,h),label='Derivada por defincion')
plt.scatter(x,cuartaDev(x),label='Derivada Exacta')
plt.legend()
plt.title("Grafica de ambos metodos de derivacion")
```

```python
plt.scatter(x,abs(cuartaDev(x)-DervidadPorDef(f,x,h)),label='Diferencia de ambas graficas')
plt.legend()
plt.title("Error del metodo de derivacion")
```
